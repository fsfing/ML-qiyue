```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y) 
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny, color="b",linestyle=":")
plt.plot(x,cosy, color="r",linestyle="--")
plt.show()
#  ":",  "-.",  "--",  "-"
```

```python
plt.plot(x,siny, color="b")
plt.plot(x,cosy, color="r")
plt.xlim(-10,15)
plt.ylim(-2,2)
plt.show()
```

```python
plt.plot(x,siny, color="b")
plt.plot(x,cosy, color="r")
plt.axis([-1,11,-1.5,1.5]) # x
plt.show()
```

```python
plt.plot(x,siny, color="b")
plt.plot(x,cosy, color="r")
plt.axis([-1,11,-1.5,1.5]) 
plt.xlabel("x number")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x,siny, color="b",label="sin{x}")
plt.plot(x,cosy, color="r",label="cos{x}")
plt.axis([-1,11,-1.5,1.5]) 
plt.xlabel("x number")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
plt.plot(x,siny, color="b",label="sin{x}")
plt.plot(x,cosy, color="r",label="cos{x}")
plt.axis([-1,11,-1.5,1.5]) 
plt.xlabel("x number")
plt.ylabel("y value")
plt.legend()
plt.title("learning the MC")
plt.show()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
x = iris.data[:,:2]
x.shape
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
y = iris.target
plt.scatter(x[y==0,0],x[y==0,1],color="red",marker="o")
plt.scatter(x[y==1,0],x[y==1,1],color="blue",marker="*")
plt.scatter(x[y==2,0],x[y==2,1],color="green",marker="+")
plt.show()
```

```python
xx = iris.data[:,2:]
yy = iris.target
plt.scatter(xx[yy==0,0],xx[yy==0,1],color="red",marker="o")
plt.scatter(xx[yy==1,0],xx[yy==1,1],color="blue",marker="*")
plt.scatter(xx[yy==2,0],xx[yy==2,1],color="green",marker="+")
plt.show()
```
