### 从sklearn 中 导入鸢尾花的数据

```python
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris()
x = iris.data
y = iris.target
```

```python
x[:10,:]
```

### 把导入的数据集分割成，测试数据集和训练数据集

```python
from playML.model_selection import train_test_split
x_train,x_test,y_train,y_test  = train_test_split(x,y,test_ratio=0.2,seed=777)
```

### 导入KNN分类器

```python
from playML.KNN2 import KNNclassifiers
```

### 导入自己封装的 均值方差归一化

```python
from playML.preprocessing import StandardScaler
```

```python
standardScaler = StandardScaler()
```

```python
standardScaler.fit(x_train) # 对训练数据集的数据，求均值方差
```

```python
# 测试 code
standardScaler.mean_
```

```python
standardScaler.transform(x_train) # 将x根据这个StandardScaler进行均值方差归一化处理
```

```python
x_train = standardScaler.transform(x_train)
x_train
```

```python
x_test_standard = standardScaler.transform(x_test) # 测试数据集，归一化
x_test_standard
```

```python
my_knn_clf = KNNclassifiers(k=3) # 系数 
```

```python
my_knn_clf.fit(x_train,y_train) # 对训练数据集，进行拟合
```

```python
my_knn_clf.score(x_test_standard, y_test) # 分类准确度
```
